```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
transactions = pd.read_csv(r'D:\Machine_Learning\data\raw\transactions_large.csv')
customers = pd.read_csv(r'D:\Machine_Learning\data\raw\customers_large.csv')
```

```python
transactions['visit_date'] = pd.to_datetime(transactions['visit_date'])
```

```python
transactions['year_month'] = transactions['visit_date'].dt.to_period('M')
```

```python
avg_spending_per_month = (
    transactions
    .groupby('year_month')['total_spending']
    .mean())
```

```python
avg_spending_per_month.plot(kind='bar', figsize=(12, 5))
plt.title('Average Spending per Month')
plt.xlabel('Month')
plt.ylabel('Average Spending')
plt.xticks(rotation=45)
plt.show()
```

```python
customers_per_city = (customers
                      .groupby('city')['customer_id']
                      .nunique())
```

```python
customers_per_city.plot(kind='bar', figsize=(12, 5))
plt.title('Number of Unique Customers per City')
plt.xlabel('City')
plt.ylabel('Customers')
plt.xticks(rotation=45)
plt.show()
```
